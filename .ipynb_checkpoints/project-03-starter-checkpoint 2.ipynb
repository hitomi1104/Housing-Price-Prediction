```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score,cross_val_predict, train_test_split, KFold, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, RidgeCV, LassoCV, Ridge, LogisticRegression,Lasso, ElasticNet, ElasticNetCV
from sklearn import metrics
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# from sklearn.datasets import load_housing
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split, cross_val_score, cross_val_predict
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import KFold
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
```

```python
#creating training and test split, pre-2010 and 2010 data splits

X = house_fixed_dum.copy()
y = X.pop('SalePrice')
```

```python
X_train = X[house_fixed_dum['YrSold']<2010]
X_test = X[house_fixed_dum['YrSold']==2010]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=1)
```

```python
# standardlize the train/test sets of data
scaler = StandardScaler()
X_train = pd.DataFrame(scaler.fit_transform(X_train), 
                       columns=X_train.columns)
X_test = pd.DataFrame(scaler.transform(X_test), 
                      columns=X_train.columns)
```

```python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
# function that returns cv score(5), training, and test scores
def model_fitter(model, X_train, y_train, X_test, y_test, cv=5):
    model.fit(X_train, y_train)
    cv_score = np.mean(cross_val_score(model, X_train, y_train, cv=cv))
    training_score = model.score(X_train, y_train)
    test_score = model.score(X_test, y_test)
    return cv_score, training_score, test_score
```

### Fit a linear regression on the training set

```python
model = LinearRegression()
model_fitter(model, X_train, y_train, X_test, y_test)
```

```python
predictions1 = model.predict(X_test)
sns.set_style("darkgrid")
plt.figure(figsize=(12,10))
sns.regplot(predictions1,y_test)
```

```python
# Checking errors
print(mean_absolute_error(y_test,predictions1))
print(mean_squared_error(y_test,predictions1))
print(np.sqrt(mean_squared_error(y_test,predictions1)))
```

### K-Fold cross-validation demonstration and R2 Scores

```python
# Perform 5-fold cross validation
scores = cross_val_score(model, X, y, cv=5)
print("Cross-validated scores:", scores)
print("Mean of Ccoss-validated scores:", scores.mean())


# Make cross validated predictions on the test sets
predictions = cross_val_predict(model, X, y, cv=5)
plt.figure(figsize=(6, 6))
y_range = [y.min(), y.max()]
plt.plot(y_range, y_range, lw=2, c='k')
plt.scatter(predictions, y)
plt.xlabel(r'$\hat{y}$')
plt.ylabel('y')
plt.show()
```

```python
#regularising data
alpha_vals = np.logspace(-3,3,50)
ridge_cv = RidgeCV(alphas = alpha_vals, cv=3)
ridge_cv.fit(X_train,y_train)
best_alpha1 = ridge_cv.alpha_
best_alpha1
```

### Fit a Ridge regression on the training set

```python
model = Ridge(alpha=1)
model_fitter(model, X_train, y_train, X_test, y_test)
```

```python
best_ridge = Ridge(alpha=best_alpha1)
cross_val_score(best_ridge,X_train,y_train,cv=3).mean()
```

### Fit Rasso on the training set

```python
model = Lasso(alpha=0.1)
model_fitter(model, X_train, y_train, X_test, y_test)
```

```python
lasso_cv = LassoCV(alphas=alpha_vals,cv=3)
lasso_cv.fit(X_train,y_train)
best_alpha2 = lasso_cv.alpha_
best_alpha2

best_lasso = Lasso(alpha=best_alpha2)
cross_val_score(best_lasso,X_train,y_train,cv=3).mean()
```

### Fit ElasticNet on the training set

```python
model = ElasticNet(alpha=0.1, l1_ratio=0.05)
model_fitter(model, X_train, y_train, X_test, y_test)
```

```python
elasticnet_cv = ElasticNetCV(alphas=alpha_vals,cv=3)
elasticnet_cv.fit(X_train,y_train)
best_alpha3 = elasticnet_cv.alpha_
best_alpha3

best_elasticnet = ElasticNet(alpha=best_alpha3)
cross_val_score(best_elasticnet,X_train,y_train,cv=3).mean()
```
